```python
import numpy as np
import pandas as pd
import scanpy as sc
import squidpy as sq
import scipy.io
import anndata
```

```python
def filter_genes_cells(adata_sc, gene_threshold_percent=1, cell_threshold_percent=1):
    non_zero_expression_genes = (adata_sc.X > 0).sum(axis=0).flatten()
    n_cells = adata_sc.shape[0]
    gene_threshold = (gene_threshold_percent / 100) * n_cells
    genes_to_keep = non_zero_expression_genes >= gene_threshold
    
    non_zero_expression_cells = (adata_sc.X > 0).sum(axis=1).flatten()
    cell_threshold = (cell_threshold_percent / 100) * adata_sc.shape[1]
    cells_to_keep = non_zero_expression_cells >= cell_threshold
    
    adata_sc_filtered = adata_sc[cells_to_keep, genes_to_keep]
    return adata_sc_filtered

output_dir = './dataset/'
adata_st = sc.read_h5ad(f'{output_dir}/raw_STdata.h5ad')
adata_sc = sc.read_h5ad(f'{output_dir}/raw_scRNAdata.h5ad')
adata_sc_filtered = filter_genes_cells(adata_sc, gene_threshold_percent=1, cell_threshold_percent=1)
adata_st_filtered = filter_genes_cells(adata_st, gene_threshold_percent=1, cell_threshold_percent=1)

adata_st_filtered.write(f'{output_dir}/STdata.h5ad')
adata_sc_filtered.write(f'{output_dir}/scRNAdata.h5ad')
```
